```python
import matplotlib.pyplot as pp
```

# LINE PLOT

```python
pp.plot([1,2,3,4,5,6,7],[3,7,1,8,2,4,5])
pp.axis([0,8,0,8])
pp.show()
```

# SCATTER PLOT

```python
x1=[1,2,3]
y1=[2.5,7,9]
x2=[2,4,6]
y2=[1,3,5]
pp.scatter(x1,y1,label='A',color='r')
pp.scatter(x2,y2,label='B',color='g')
pp.title("Scatter Plot",color='b')
pp.xlabel('X')
pp.ylabel('Y')
pp.legend()
pp.show()
```

# HISTOGRAM

```python
d=[10,23,31,42,54,65,76,89,90,105]
pp.hist(d,color='r')
pp.title('Histogram')
pp.show()
```

# BAR PLOT

```python
pp.bar([1,2,3,4,5,6],[10,20,30,40,50,70],label='A',color='DarkBlue',width=0.5)
pp.bar([1,2,3,4,5,6],[20,10,30,20,40,60],label='B',color='LightGreen',width=0.25)
pp.title('Bar Plot',color='DarkGreen')
pp.xlabel('X')
pp.ylabel('Y')
pp.legend()
pp.show()
```

# AREA PLOT

```python
DAY=[0,1,2,3,4,5]
age=[63,81,43,23,98,65]
weight=[57,45,48,60,52,40]
pp.plot([],[],label='A',color='c',linewidth=15)
pp.plot([],[],label='B',color='b',linewidth=10)
pp.stackplot(DAY,age,weight,colors=['c','b'])
pp.legend()
pp.show()
```

# PIE CHART

```python
slice=[10,23,45,34,12]
act=['harish','santhosh','nivash','naveen','monish']
c=['r','g','b','orange','y']
pp.pie(slice,labels=act,colors=c,startangle=90,shadow=False,explode=(0.2,0.2,0.1,0.2,0.2),autopct='%1.1f%%')
pp.show()
```
